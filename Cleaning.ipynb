```python
def parse_ram(s):
    gb_match = re.search(r'(\d+)\s*GB?(?!\w)', s, re.IGNORECASE)
    ddr_match = re.search(r'DDR[3-9]', s, re.IGNORECASE)
    
    gb = gb_match.group(1) if gb_match else None
    ddr = ddr_match.group(0).upper() if ddr_match else None
    
    if gb and ddr:
        return f"{gb}GB {ddr}"
    elif gb:
        return f"{gb}GB"
    return None
```

```python
def extract_storage_clean(text):
    # Remove frequencies, MHz, MT/s
    cleaned = re.sub(r'\d+(?:\.\d+)?\s*GHz', '', text, flags=re.IGNORECASE)
    cleaned = re.sub(r'\d+\s*MHz', '', cleaned, flags=re.IGNORECASE)
    cleaned = re.sub(r'\d+\s*MT/s', '', cleaned, flags=re.IGNORECASE)

    # Storage type: optional PCIe/Gen prefix + storage keyword
    storage_type = (
        r'(?:'
            r'(?:PCIe|PCle|PCI-E)\s*(?:Gen\s*\d+\s*)?(?:\d+(?:\.\d+)?x?\d*)?\s*'  # PCIe 4.0 / PCIe Gen4
            r'|(?:Gen\s*\d+\s*)'                                                      # standalone Gen4 / Gen 4
        r')?'
        r'(?:NVMe|NVME|M\.2|HDD|SSD|Storage|Solid\s*State\s*Drive)'
    )

    pattern = rf'(\d+(?:\.\d+)?)\s*(TB|GB|T|G)\s*x?\d*\s*({storage_type})'

    matches = re.findall(pattern, cleaned, re.IGNORECASE)

    unit_map = {'T': 'TB', 'G': 'GB', 'TB': 'TB', 'GB': 'GB'}

    results = []
    for size, unit, stype in matches:
        unit_clean = unit_map.get(unit.upper(), unit.upper())
        stype_clean = re.sub(r'\s+', ' ', stype.strip())
        results.append(f"{size}{unit_clean} {stype_clean}")

    # Fallback: standalone NVME with no adjacent size
    if not results and re.search(r'\bNVME\b', cleaned, re.IGNORECASE):
        size_match = re.search(r'(\d+)\s*(TB|GB|T|G)', cleaned, re.IGNORECASE)
        if size_match:
            size, unit = size_match.groups()
            unit_clean = unit_map.get(unit.upper(), unit.upper())
            results.append(f"{size}{unit_clean} NVMe")

    return results[-1] if results else None
```

```python
def clean_os(row):
    if pd.isna(row):
        return None

    s = str(row).upper()
    if 'W-11' in s or 'WIN 11' in s or 'WINDOW 11' in s or 'WINDOWS 11' in s or 'WIN11' in s:
        if 'PRO' in s:
            return 'Windows 11 Pro'
        elif 'HOME' in s:
            return 'Windows 11 Home'
        else:
            return 'Windows 11'

    elif 'WINDOWS 10' in s or 'WIN 10' in s:
        if 'PRO':
            return 'Windows 10 Pro'
        elif 'HOME':
            return 'Windows 10 Home'
        else:
            return 'Windows 10'

    return None
```
