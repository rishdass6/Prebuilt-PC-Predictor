```python
X = df_clean.iloc[:, :-1]
y = df_clean.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 0)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import TargetEncoder, StandardScaler 
from sklearn.linear_model import RidgeCV
from sklearn.ensemble import (RandomForestRegressor, GradientBoostingRegressor, StackingRegressor)
from sklearn.model_selection import cross_val_score

cat_cols = X_train.select_dtypes(include = 'object').columns.tolist()
num_cols = X_train.select_dtypes(exclude = 'object').columns.tolist()
print(cat_cols)
print(num_cols)
```

```python
from sklearn.compose import ColumnTransformer

col_encoder = ColumnTransformer(transformers = [
     ('target_enc', TargetEncoder(), cat_cols),
     ('num', StandardScaler(), num_cols)
])
```

```python
y_log = np.log1p(y_train)

ridge_model = Pipeline([
    ('Preprocess', col_encoder),
    ('model', RidgeCV(alphas = [0.01, 1, 10, 100]))
])

gb_model = Pipeline([
    ('Preprocess', col_encoder),
    ('Model', GradientBoostingRegressor(n_estimators = 200, max_depth = 3, learning_rate = 0.05))
])

stack_model = Pipeline([
    ('Preprocess', col_encoder),
    ('Model', StackingRegressor(
        estimators = [
            ('ridge', RidgeCV(alphas = [0.1, 0.5, 1, 10, 100])), 
            ('rf', RandomForestRegressor(n_estimators = 200, max_depth = 20, random_state = 42))
        ],
        final_estimator = RidgeCV()
    ))
])
```

```python
models = {
    'Ridge': ridge_model,
    'GradientBoost': gb_model,
    'StackingRegressor': stack_model
}

for name, model in models.items():
    score = cross_val_score(model, X_train, y_log, cv = 10, scoring = 'r2')
    print(f'{name} - Mean: {score.mean().round(4)}, Std: {score.std().round(4)}')
```

```python
best_model = stack_model

best_model.fit(X_train, y_log)
y_pred_log = best_model.predict(X_test)
y_pred_actual = np.expm1(y_pred_log)

from sklearn.metrics import r2_score, mean_absolute_error
print("Test R²:", r2_score(np.expm1(np.log1p(y_test)), y_pred_actual))
print("MAE:", mean_absolute_error(y_test, y_pred_actual))
```

```python
estimators = [
    ('ridge', Ridge(alpha = 1)),
    ('rf', RandomForestRegressor(n_estimators = 1000, min_samples_leaf = 1, max_features = 'sqrt', max_depth = 30)),
    ('gb', GradientBoostingRegressor(n_estimators = 300, max_depth = 3, learning_rate = 0.05))
]

stack_model = Pipeline([
    ('Preprocess', col_encoder),
    ('Model', StackingRegressor(
        estimators = estimators,
        final_estimator = RidgeCV(),
        cv = 5
    ))
])

scores = cross_val_score(stack_model, X_train, np.log1p(y_train), cv = 10, scoring = 'r2')
print('Mean: ', scores.mean().round(4))
print('Std: ', scores.std().round(4)) 
```
