```python
df['CPU'].value_counts().sort_index().index.tolist()
```

```python
cpu_pattern = (
    r'^(?P<company>AMD|Intel)\s+'
    r'(?P<family>Ryzen\s+[0-9]|Ryzen|Xeon|Ultra\s+[0-9]|i[3-9])'
    r'(?:\s+(?P<generation>[0-9A-Z]+))?'
)
```

```python
df[['CPU_COMPANY', 'CPU_MODEL', 'CPU_GEN']] = df['CPU'].str.extract(cpu_pattern)
```

```python
gpu_pattern = (
    r'^(?P<company>AMD|NVIDIA|Intel)\s+'
    r'(?P<label>RTX|RX|GTX|Arc|Vega)\s+'
    r'(?P<model>[B0-9][0-9]*[0-9A-Za-z]*)\s*'
    r'(?P<suffix>Ti|Super|XT)?$'
)
```

```python
df[['GPU_COMPANY', 'GPU_LABEL', 'GPU_MODEL', 'GPU_SUFFIX']] = df['GPU'].str.extract(gpu_pattern)
```

```python
df
```

```python
df['GPU_LABEL'].isna().sum()
```
